```python
# read the csv file 
PROCESSED_DIR = Path.cwd().parent / "data" / "processed"

experiment1_= PROCESSED_DIR / "experiment1_output.csv"

```

```python
df = pd.read_csv(experiment1_)
df.sample(1)
```

```python
import ast 
def convert_ast(obj):
        """Converts string representation of list to actual list."""
        if isinstance(obj, str):
            return ast.literal_eval(obj)
        return obj
```

```python
df1 = df.copy()
```

```python
columns_to_convert = ['genres', 'overview', "production_companies","production_countries" , 'keywords', 'cast', 'crew']

for col in columns_to_convert:
    df1[col] = df1[col].apply(convert_ast)

```
