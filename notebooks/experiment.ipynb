```python
# @title
import ast

def get_primary_spoken_language_iso(languages_str):
    """Extracts the ISO_639_1 code of the first spoken language from the spoken_languages string."""
    if pd.isna(languages_str) or languages_str == '[]':
        return None
    try:
        languages = ast.literal_eval(languages_str)
        if languages and isinstance(languages, list) and 'iso_639_1' in languages[0]:
            return languages[0]['iso_639_1']
    except (ValueError, SyntaxError):
        pass # Handle cases where string is not a valid list representation
    return None

# Create a temporary column with primary spoken language ISO codes
movie_metadata['primary_spoken_language_iso'] = movie_metadata['spoken_languages'].apply(get_primary_spoken_language_iso)

# Fill missing original_language values using the temporary column
# Only apply to rows where original_language is NaN
movie_metadata['original_language'].fillna(movie_metadata['primary_spoken_language_iso'], inplace=True)

# Drop the temporary column
movie_metadata.drop(columns=['primary_spoken_language_iso'], inplace=True)

print(f"Number of null values in 'original_language' after filling: {movie_metadata['original_language'].isnull().sum()}")
print(movie_metadata[movie_metadata.original_language.isna()][['original_title', 'spoken_languages', 'original_language']])
```
